```python
# alps = []
# mean = []
# length= []
# std_ratios = []
for alpha in range(30,40):
    alpha = alpha/100
    for std_ratio in range(1,10):
#     for std_ratio in [0]:
        std_ratio = std_ratio/100
        # alps = []
        # mean = []
        # for alpha in range(100):
    #     alpha = 0.45
        #compare the p(t) with p(t-tau)
        S1 = []
        S2 = []
        predictions = []
        true = []
        tau =1
        t = 0

        error = []
        std = []
        for row in data[['betaTrue',columns[0],columns[1]]].iterrows():
            spread = row[1].values[0]
            assetA = row[1].values[1]
            assetB = row[1].values[2]
            
            if len(predictions)>0:     
                predictedSpread = assetA-predictions[-1]*assetB

                error.append(predictedSpread)
                std.append(np.std(error))
            else:
                error.append(0)
                std.append(0)

            if len(predictions)>0:
                true.append(spread)


            if len(S1)==0:
                S1.append(spread)
                S2.append(spread)
            else:
                S1.append(alpha*spread + (1-alpha)*S1[-1])
                S2.append(alpha*S1[-1] + (1-alpha)*S2[-1])



            prediction = (2+ alpha*tau/(1-alpha))*S1[-1] - (1+alpha*tau/(1-alpha))*S2[-1]
            #Prediction for the next time step
            predictions.append(prediction)
            #true.append(spread at next time step)

        #match the length of array and predictions
        
        #beta is x/y
        #predictions are x - beta * y
        
        #ignore p(n+1)th prediction
        predictions = predictions[:-1]
        #append 0 so it fits dataframe
        predictions.append(0)
        true.append(0)
        data.loc[slice(None),'true']= true
        data.loc[slice(None),'predicted'] = predictions

        data.loc[slice(None),'accuracy']= data['true']-data['predicted']


        data.loc[slice(None),'error'] = error
        data.loc[slice(None),'std'] = std
        
        #Adjust harshness of entry
        data.loc[slice(None),'std'] =  data['std']*std_ratio 
            
        data.loc[slice(None),'longSignal'] = data['error'] > data['std']
        data.loc[slice(None),'longExit'] = data['error'] < data['std']
        data.loc[slice(None),'shortSignal'] = data['error'] <- data['std']
        data.loc[slice(None),'shortExit'] = data['error'] > - data['std']

        data.loc[slice(None),'short'] = 0
        data.loc[slice(None),'short'][data['shortSignal']] = -1
        data.loc[slice(None),'short'][data['shortExit']] = 0
        data.loc[slice(None),'long'] = 0
        data.loc[slice(None),'long'][data['longSignal']] = 1
        data.loc[slice(None),'long'][data['longExit']] = 0

        data.loc[slice(None),'numUnits'] = data['long'] + data['short']

        data.loc[slice(None),'numUnits'+columns[0]] =  -data['numUnits']
        data.loc[slice(None),'numUnits'+columns[1]] =  data['numUnits']*data['predicted']

        data.loc[slice(None),'positions'+columns[0]] = data['numUnits'+columns[0]]*data[columns[0]]
        data.loc[slice(None),'positions'+columns[1]] = data['numUnits'+columns[1]]*data[columns[1]]

        # comission = (data['positions'+columns[0]].shift(1)+data['positions'+columns[1]].shift(1))*0.0008
        comission = 0
        data.loc[slice(None),'pnl'] = data[columns[0]].pct_change()*(data['positions'+columns[0]].shift(1)) + data[columns[1]].pct_change()*(data['positions'+columns[1]].shift(1))-comission




        data.loc[slice(None),'returns'] = (data['pnl'])/(np.abs(data['positions'+columns[0]]) +np.abs(data['positions'+columns[0]])).shift(1)
        alps.append(alpha)
        mean.append(np.cumsum(data['returns'].fillna(0)).iloc[-1])
        length.append(len(data.returns.dropna()))
        std_ratios.append(std_ratio)
```
